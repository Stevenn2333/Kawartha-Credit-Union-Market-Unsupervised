## Data Description
Source: [Bank Marketing](https://archive.ics.uci.edu/ml/datasets/bank+marketing)

The data is related with direct marketing campaigns of a Portuguese banking institution. The marketing campaigns were based on phone calls. Often, more than one contact to the same client was required. 

#### Importance
The dataset being used in our solution comprises bank client data, including age, job type, marital status, education level, credit default status, average yearly balance, housing loan status, personal loan status, and other attributes related to the last contact of the current marketing campaign, such as the contact communication type, last contact day of the month, last contact month of the year, and last contact duration. Additionally, the dataset includes information about the number of contacts performed during this campaign and for this client, the number of days that passed after the client was last contacted from a previous campaign, the number of contacts performed before this campaign and for this client, and the outcome of the previous marketing campaign. Our desired output is whether or not the client has subscribed to a term deposit. This dataset will be used to train a model that predicts the likelihood of a client subscribing to a term deposit, which can be used by our solution to make data-driven decisions about marketing strategies and client outreach.

### Input variables:
#### Bank client data:
1. age (numeric) 
2. job : type of job (categorical: "admin.","unknown","unemployed","management","housemaid","entrepreneur","student","blue-collar","self-employed","retired","technician","services") 
3. marital : marital status (categorical: "married","divorced","single"; note: "divorced" means divorced or widowed)
4. education (categorical: "unknown","secondary","primary","tertiary")
5. default: has credit in default? (binary: "yes","no")
6. balance: average yearly balance, in euros (numeric) 
7. housing: has housing loan? (binary: "yes","no")
8. loan: has personal loan? (binary: "yes","no")
#### Related with the last contact of the current campaign:
9. contact: contact communication type (categorical: "unknown","telephone","cellular") 
10. day: last contact day of the month (numeric)
11. month: last contact month of year (categorical: "jan", "feb", "mar", ..., "nov", "dec")
12. duration: last contact duration, in seconds (numeric)
#### Other attributes:
13. campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)
14. pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric, -1 means client was not previously contacted)
15. previous: number of contacts performed before this campaign and for this client (numeric)
16. poutcome: outcome of the previous marketing campaign (categorical: "unknown","other","failure","success")

#### Suppose Output (desired target):
17. has the client subscribed a term deposit? (binary: "yes","no")

## Training

### Set Up

#### K Means Clustering Training for K
Cluster training using k means is performed by comparing the performance of different k values, plotted on an elbow plot. Where the elbow is seen indicates a dip in performance with respect to k value. An ideal k value is identified by examining this elbow plot. This identified k value will be used for the actual unsupervised learning model.

`````python
%pip install mxnet
`````

`````python
import time
from time import gmtime, strftime
from sagemaker.amazon.amazon_estimator import get_image_uri

# Get the K-Means docker image from AWS
image = get_image_uri(region, 'kmeans')

# Get the paths for training, output, and validation data
data_location = training_data_path
output_location = output_data_path
test_location = testing_data_path

# Set unique names for the output so that it can be differentiated from previous output
output_time = strftime("%Y-%m-%d-%H-%M-%S", gmtime())
output_folder = 'kmeans-k-discovery-' + output_time

# Specify configurations for exploring k values
K = range(2, 12) # change the range to be used for k
INSTANCE_COUNT = 2
run_parallel_jobs = True 

job_names = []

# Perform the training job on each value of k
for k in K:
    print('Starting train job: ' + str(k))
    output_location = 's3://{}/Bank/kmeans-k-discovery/output/'.format(bucket) + output_folder
    print('Training artifacts will be uploaded to: {}'.format(output_location))
    job_name = output_folder + str(k)

    create_training_params = {
        "AlgorithmSpecification": {
            "TrainingImage": image,
            "TrainingInputMode": "File"
        },
        "RoleArn": role,
        "OutputDataConfig": {
            "S3OutputPath": output_location
        },
        "ResourceConfig": {
            "InstanceCount": INSTANCE_COUNT,
            "InstanceType": "ml.m5.large",
            "VolumeSizeInGB": 50
        },
        "TrainingJobName": job_name,
        "HyperParameters": {
            "k": str(k),
            "feature_dim": "13",
            "mini_batch_size": "1000"
        },
        "StoppingCondition": {
            "MaxRuntimeInSeconds": 60 * 60
        },
                'InputDataConfig': [
            {
                'ChannelName': 'train',
                'DataSource': {
                    'S3DataSource': {
                        'S3DataType': 'S3Prefix',
                        'S3Uri': data_location,
                        'S3DataDistributionType': 'FullyReplicated'
                    }
                },
                'ContentType': 'text/csv;label_size=0'
            },
            
            {
                'ChannelName': 'test',
                'DataSource': {
                    'S3DataSource': {
                        'S3DataType': 'S3Prefix',
                        'S3Uri': test_location,
                        'S3DataDistributionType': 'FullyReplicated'
                    }
                },
                'ContentType': 'text/csv;label_size=0'
            }
        ]
    }

    # Create the training job on the SageMaker runtime, using the training parameters seen above for the job
    runtime.create_training_job(**create_training_params)

    # Print the status of the training job
    status = runtime.describe_training_job(TrainingJobName=job_name)['TrainingJobStatus']
    print(status)
    
    job_names.append(job_name)
`````

`````python
# Go through each of the training jobs, and report its status
while len(job_names):
    try:
        runtime.get_waiter('training_job_completed_or_stopped').wait(TrainingJobName=job_names[0])
    finally:
        status = runtime.describe_training_job(TrainingJobName=job_name)['TrainingJobStatus']
        print("Training job ended with status: " + status)
        if status == 'Failed':
            message = runtime.describe_training_job(TrainingJobName=job_name)['FailureReason']
            print('Training failed with the following error: {}'.format(message))
            raise Exception('Training job failed')

    info = runtime.describe_training_job(TrainingJobName=job_name)
    job_names.pop(0)
`````

### Hyperparameter Tuning
Hyperparameter tuning involves specifying an objective metric and other training parameters. Using default values for the rest of the settings that will spcify the number of iterations, instances, and other testing parameters, the objective metric is one of the most important features of the hyperparameter tuning. The objective measure that is going to be used is `test:ssd`, which represents the sum of the squared distances between the records and their nearest centroid. The metric is used in model training by minimizing this value in the test data.

`````python
# The ideal k value, obtained from the elbow plot above
k = "5"

# Set the job name using the current date and time, to be differentiated from other models
tuning_job_name = "bank-data" + strftime("%Y-%m-%d-%H-%M-%S", gmtime())

# Set the testing, output, and validation data location paths
data_location = training_data_path
output_location = output_data_path
test_location = testing_data_path

# Create a new SageMaker client
sagemaker = boto3.client('sagemaker')

# Create the hyperparameter tuning job, specifying the config to be used
response = sagemaker.create_hyper_parameter_tuning_job(
    HyperParameterTuningJobName= tuning_job_name,
    HyperParameterTuningJobConfig={
        'Strategy': 'Bayesian',
        'HyperParameterTuningJobObjective': {
            'Type': 'Minimize',
            'MetricName': 'test:ssd'
        },
        'ResourceLimits': {
            'MaxNumberOfTrainingJobs': 10,
            'MaxParallelTrainingJobs': 2
        },
        'ParameterRanges': {
            'IntegerParameterRanges': [
                {
                    'Name': 'mini_batch_size',
                    'MinValue': '3000',
                    'MaxValue': '15000'
                },
            ],
            'CategoricalParameterRanges': [
                {
                    'Name': 'init_method',
                    'Values': [
                        'kmeans++', 'random'
                    ]
                },
            ]
        },
        'TrainingJobEarlyStoppingType' : 'Auto'
    },
    TrainingJobDefinition={
        'StaticHyperParameters': {
            'k': k,
            'feature_dim': '13',
        },
        'AlgorithmSpecification': {
            'TrainingImage': image,
            'TrainingInputMode': 'File'
        },
        'RoleArn': role,
        'InputDataConfig': [
            {
                'ChannelName': 'train',
                'DataSource': {
                    'S3DataSource': {
                        'S3DataType': 'S3Prefix',
                        'S3Uri': data_location,
                        'S3DataDistributionType': 'FullyReplicated'
                    }
                },
                'ContentType': 'text/csv;label_size=0'
            },
            
            {
                'ChannelName': 'test',
                'DataSource': {
                    'S3DataSource': {
                        'S3DataType': 'S3Prefix',
                        'S3Uri': test_location,
                        'S3DataDistributionType': 'FullyReplicated'
                    }
                },
                'ContentType': 'text/csv;label_size=0'
            },
        ],
        'OutputDataConfig': {
                'S3OutputPath': output_location
        },
        'ResourceConfig': {
            'InstanceType': 'ml.m4.xlarge',
            'InstanceCount': 1,
            'VolumeSizeInGB': 50
        },
        'StoppingCondition': {
            'MaxRuntimeInSeconds': 60 * 60
        }
    }
)

# Perform the tuning job using the config detailed above
tuning_job_result = sagemaker.describe_hyper_parameter_tuning_job(
    HyperParameterTuningJobName=tuning_job_name
)

status = tuning_job_result["HyperParameterTuningJobStatus"]

# Wait until the tuning job is complete, checking periodically until it is complete
while status != "Completed":
    tuning_job_result = sagemaker.describe_hyper_parameter_tuning_job(
        HyperParameterTuningJobName=tuning_job_name
    )
    status = tuning_job_result["HyperParameterTuningJobStatus"]
    print(status)

    job_count = tuning_job_result["TrainingJobStatusCounters"]["Completed"]
    print("Number of jobs completed", job_count)
    
    time.sleep(60)

# Display the results of the tuning job
objective = tuning_job_result["HyperParameterTuningJobConfig"]["HyperParameterTuningJobObjective"]
is_minimize = objective["Type"] != "Maximize"
objective_name = objective["MetricName"]
`````

#### Make the Lambda Function
Create the function, author from scratch, name it, choose Python for runtime, change execution role to LabRole, create the function. Add the following code:

import os
import os
import io
import boto3
import json
import csv

ENDPOINT_NAME = os.environ['ENDPOINT_NAME'] # get endpoint name
runtime= boto3.client('runtime.sagemaker') # make low level sagemaker client

def lambda_handler(event, context):
    data = json.loads(json.dumps(event)) 
    payload = data['data'] #get data from event param
    print(payload)
    response = runtime.invoke_endpoint(EndpointName=ENDPOINT_NAME,
        ContentType='text/csv',
        Body=payload)  #call endpoint with endpoint name, type, and payload
    result = json.loads(response['Body'].read().decode()) # get response
    print(result)
    return {
        'statusCode': 200,
        'headers': {
            'Access-Control-Allow-Headers': 'Content-Type',
            'Access-Control-Allow-Origin': '*',
            'Access-Control-Allow-Methods': 'OPTIONS,POST,GET'
        },
        'body': result
    }
    
The print statements in the Lambda help verify that the calls are going through correctly. Under configuration add the endpoint name under ENVIRONMENT_VARIABLE, and then deploy the Lambda function.
